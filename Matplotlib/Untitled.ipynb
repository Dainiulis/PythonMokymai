```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = x * 2

x2 = np.arange(0, 4.5, 0.5)
plt.figure(figsize=(5,3), dpi=300)
plt.plot(x2, x2**2, 'y.--', linewidth=2, label='**2')
plt.plot(x, y, color='red', label='2x', linewidth=3)
plt.title('Pirmas grafikas', fontdict={'fontname': 'Comic Sans MS', 
                                       'fontsize': 20})
plt.xlabel('x asis')
plt.ylabel('y asis')

plt.legend()

plt.show()
```
