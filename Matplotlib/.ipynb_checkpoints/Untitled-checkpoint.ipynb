```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x, y)
plt.title('Pirmas grafikas', fontdict={'fontname': 'Comic Sans MS', 
                                       'fontsize': 20})
plt.xlabel('x asis')
plt.ylabel('y asis')
plt.show()
```
