## Užduotis 

- Pateiktas Iris duomenų rinkinys- vienas iš klasikinių duomenų rinkinių algoritmams pasibandyti. Daugiau informacijos apie rinkinį: https://archive.ics.uci.edu/ml/datasets/iris

- Duomenų rinkinys - informacija apie Iris (lietuviškai - Vilkdalgio) augalus. Kiekvienas duomenų rinkinio elementas susideda iš keturių savybių apibūdinančių augalą. Kiekvienas augalas gali priklausyti vienai iš trijų klasių (plačiau - žiūrėti pridėtą nuorodą)

- Jūsų užduotis- pamėginti klasterizuoti pateiktus duomenis, pateikti savo įžvalgas.

```python
# nesiūloma taikyti PCA

from sklearn.cluster import KMeans
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

Po kiekvienu algoritmu yra komentaras

```python
# indeksas nurodo kurią reikšmę pasiimti
colors = ['red', 'green', 'yellow']
labels = ['Iris_Setosa', 'Iris_Versicolour', 'Iris_Virginica']
```

```python
print("Data shape: ", X.shape)
print("Labels shape: ", y.shape)
print(X[:10, 2:])
print(y[:10])
print(np.unique(y))
```

```python
# sepal and petal real data
def plot_original_data():
    fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(15,5))
    for i in np.unique(y):
        rows = np.where(y == i)
        ax0.scatter(X[rows, 0], X[rows, 1], color=colors[i], label=labels[i])
        ax0.legend()
        ax0.set_title('sepal')
        ax1.scatter(X[rows, 2], X[rows, 3], color=colors[i], label=labels[i])        
        ax1.legend()
        ax1.set_title('petal')
    fig.suptitle('Originalūs duomenys')
    plt.show()
```

```python
plot_original_data()
```

```python
plot_original_data()

#K-Means
from sklearn.cluster import KMeans

# parenkami 3 klasteriai, nes žinoma, kad yra 3 klasės
model = KMeans(n_clusters=3)

#"Mokome"
model.fit(X)

#Atliekame spėjimus
y_pred = model.predict(X)

#Išrenkame klasterius
clusters = np.unique(y_pred)
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(15,5))
fig.suptitle('K-means')
diffs_for_cluster = {}
for cluster in clusters:
    rows = np.where(y_pred == cluster)
    orig_rows = np.where(y == cluster)
    diffs_for_cluster[cluster] = X[orig_rows, :].shape[1] - X[rows, :].shape[1]
    ax0.scatter(X[rows, 0], X[rows, 1], color=colors[cluster], label=labels[cluster])
    ax0.legend()
    ax0.set_title('sepal')
    ax1.scatter(X[rows, 2], X[rows, 3], color=colors[cluster], label=labels[cluster])
    ax1.legend()
    ax1.set_title('petal')
plt.show()
for k, v in diffs_for_cluster.items():
    print(f'Skirtumas tarp {labels[k]} klasteriu {k}: {v}')
```

K means algoritmas gauna ganėtinai panašius rezultatus, tik keista, kad spalvos susimaišę. Tikslumas nėra idealus, greičiausiai todėl, kad centroid-based algoritmai sunkiau susitvarko su šalia esančiais klasteriais

```python
plot_original_data()

#Agglomerative 

from sklearn.cluster import AgglomerativeClustering

# parenkami 3 klasteriai, nes žinoma, kad yra 3 klasės
model = AgglomerativeClustering(n_clusters=3)

#Atliekame spėjimus
y_pred = model.fit_predict(X)


#Išrenkame klasterius
clusters = np.unique(y_pred)

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(15,5))
fig.suptitle('Agglomerative')
diffs_for_cluster = {}
for cluster in clusters:
    rows = np.where(y_pred == cluster)
    orig_rows = np.where(y == cluster)
    diffs_for_cluster[cluster] = X[orig_rows, :].shape[1] - X[rows, :].shape[1]
    ax0.scatter(X[rows, 0], X[rows, 1], color=colors[cluster], label=labels[cluster])
    ax0.legend()
    ax0.set_title('sepal')
    ax1.scatter(X[rows, 2], X[rows, 3], color=colors[cluster], label=labels[cluster])
    ax1.legend()
    ax1.set_title('petal')

plt.show()


for k, v in diffs_for_cluster.items():
    print(f'Skirtumas tarp {labels[k]} klasteriu {k}: {v}')
```

Taip pat panašūs rezultatai kaip ir su K means algoritmu, taip pat spalvos susimaišę.
Tikslumas atrodo kiek geresnis nei K-means algoritmo

```python
plot_original_data()

#Gaussian Mixture

from sklearn.mixture import GaussianMixture

# parenkami 3 klasteriai, nes žinoma, kad yra 3 klasės
model = GaussianMixture(n_components=3)


# "Mokome"
model.fit(X)


# Atliekame spėjimus
y_pred = model.predict(X)

# Atliekame spėjimus
clusters = np.unique(y_pred)

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(15,5))
fig.suptitle('Gaussian Mixture')

diffs_for_cluster = {}
for cluster in clusters:
    rows = np.where(y_pred == cluster)
    orig_rows = np.where(y == cluster)
    diffs_for_cluster[cluster] = X[orig_rows, :].shape[1] - X[rows, :].shape[1]
    ax0.scatter(X[rows, 0], X[rows, 1], color=colors[cluster], label=labels[cluster])
    ax0.legend()
    ax0.set_title('sepal')
    ax1.scatter(X[rows, 2], X[rows, 3], color=colors[cluster], label=labels[cluster])
    ax1.legend()
    ax1.set_title('petal')

plt.show()

for k, v in diffs_for_cluster.items():
    print(f'Skirtumas tarp {labels[k]} klasteriu {k}: {v}')
```

Panašiausia į originalius duomenis. Taip yra greičiausiai todėl, kad duomenys yra pasiskirstę pagal normalųjį skirstinį

```python
plot_original_data()
# DBSCAN

from sklearn.cluster import DBSCAN


#sukuriame modelį, nurodome reikalingus hiperparametrus

model = DBSCAN(eps=0.9, min_samples=45)

# Atliekame spėjimus
y_pred = model.fit_predict(X)

#Atrenkame gautus klasterius
clusters = np.unique(y_pred)

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(15,5))
fig.suptitle('DBSCAN')

diffs_for_cluster = {}
for cluster in clusters:
    rows = np.where(y_pred == cluster)
    orig_rows = np.where(y == cluster)
    diffs_for_cluster[cluster] = X[orig_rows, :].shape[1] - X[rows, :].shape[1]
    ax0.scatter(X[rows, 0], X[rows, 1], color=colors[cluster], label=labels[cluster])
    ax0.legend()
    ax0.set_title('sepal')
    ax1.scatter(X[rows, 2], X[rows, 3], color=colors[cluster], label=labels[cluster])
    ax1.legend()
    ax1.set_title('petal')

plt.show()

for k, v in diffs_for_cluster.items():
    print(f'Skirtumas tarp {labels[k]} klasteriu {k}: {v}')
```

Gan sunku rasti tinkamus parametrus, tikslumo su originaliais duomenimis nėra, kaip su kitais algoritmais

```python
plot_original_data()

# MeanShift centroid based algoritmas
# Pats aptinka clusterius (sliding window metodas)

from sklearn.cluster import MeanShift, estimate_bandwidth

bandwidth = estimate_bandwidth(X, quantile=0.28, n_samples=X.shape[0])
print(bandwidth)
#sukuriame modelį, nurodome reikalingus hiperparametrus

model = MeanShift(bandwidth=bandwidth, bin_seeding=True)


model.fit(X)
# Atliekame spėjimus
y_pred = model.predict(X)

#Atrenkame gautus klasterius
clusters = np.unique(y_pred)
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(15,5))
fig.suptitle('MeanShift')

diffs_for_cluster = {}
for cluster in clusters:
    rows = np.where(y_pred == cluster)
    orig_rows = np.where(y == cluster)
    diffs_for_cluster[cluster] = X[orig_rows, :].shape[1] - X[rows, :].shape[1]
    ax0.scatter(X[rows, 0], X[rows, 1], color=colors[cluster], label=labels[cluster])
    ax0.legend()
    ax0.set_title('sepal')
    ax1.scatter(X[rows, 2], X[rows, 3], color=colors[cluster], label=labels[cluster])
    ax1.legend()
    ax1.set_title('petal')

plt.show()

for k, v in diffs_for_cluster.items():
    print(f'Skirtumas tarp {labels[k]} klasteriu {k}: {v}')
```
