## Sukurkite klasifikatorių, duotam duomenų rinkiniui

- Susipažinkite su duomenimis
- Pamėginkite sukurti neuroninį tinklą paveikslėliui priskiriantį vieną iš 3 galimų klasių

```python
#Įsikeliame reikalingas bibliotekas


import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
import keras

from sklearn.model_selection import train_test_split

import cv2

from pylab import rcParams
rcParams['figure.figsize'] = 10, 10
```

```python
X, y = datasets.make_classification(n_classes=3, n_samples=300, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=1)
```

```python
for i in np.unique(y):
    rows = np.where(y == i)
    if i == 0:
        color = 'red'
    elif i == 1:
        color = 'orange'
    else:
        color = 'green'
    plt.scatter(X[rows, 0], X[rows, 1], c = color)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.1, shuffle= True)
```

```python
y_train_one_hot = keras.utils.to_categorical(y_train, 3)
y_valid_one_hot = keras.utils.to_categorical(y_valid, 3)
```

```python
#Išbandykime paprastą logistinę regresiją, nenaudodami jokių vidinių paslėptų sluoksnių

model = keras.models.Sequential()
# 3 klases, todel 3 neuronai. turi 2 savybes. softmax naudojame nes > 2 klases
model.add(keras.layers.Dense(3, input_dim=2, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer=keras.optimizers.SGD(lr=0.1), metrics=['accuracy'])
model.summary()
```

```python
h = model.fit(X_train, y_train_one_hot, validation_data=(X_valid, y_valid_one_hot), epochs = 150 )
```

```python
# atvaizduojame grafiskai
plt.subplot(1, 2, 1)
plt.title('LOSS')
plt.plot(h.history['loss'], color='red')
plt.plot(h.history['val_loss'], color='green')
plt.subplot(1, 2, 2)
plt.title('Accuracy')
plt.plot(h.history['accuracy'], color='red')
plt.plot(h.history['val_accuracy'], color='green')
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Dense(30, input_dim=2, activation='relu'))
model.add(keras.layers.Dense(30, activation='relu'))
model.add(keras.layers.Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer=keras.optimizers.SGD(lr=0.1), metrics=['accuracy'])
model.summary()
```

```python
h = model.fit(X_train, y_train_one_hot, validation_data=(X_valid, y_valid_one_hot), epochs = 150, batch_size=32)
# batch size kiek per viena epocha atlieka svoriu pakeitimu
# su vaizdais galima didinti (tuomet maziau atminties naudos ir greiciau mokysis)
```

```python
# atvaizduojame grafiskai
plt.subplot(1, 2, 1)
plt.title('LOSS')
plt.plot(h.history['loss'], color='red')
plt.plot(h.history['val_loss'], color='green')
plt.subplot(1, 2, 2)
plt.title('Accuracy')
plt.plot(h.history['accuracy'], color='red')
plt.plot(h.history['val_accuracy'], color='green')
```
